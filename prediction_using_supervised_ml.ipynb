PREDICTION USING SUPERVISED ML


Predict the percentage of a student absed on the number of hours of study

```python
#Import all necessary libraries
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
#importing the data
#Here the independant variable is hours taken 
import pandas as pd 
data=pd.read_csv('http://bit.ly/w-data')
print('data imported successfully')
print('Data is shown below')
df=pd.DataFrame(data)
print(df)
X=df[['Hours']] #x=no: of hours taken 
Y=df['Scores'] #y=score achieved 
```

```python
#Plotting hours vs scores graph 
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.scatter(X,Y)
```

```python
#splitting the data for training and  testing 
#75% data used for training and 25% data used for testing 
import tensorflow as tf
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
#Linear Regression model is choosen 
from sklearn.linear_model import LinearRegression
model=LinearRegression()
#fitting the training data to the model 
model.fit(xtrain,ytrain)
```

```python
#making predictions 
predictx=model.predict(xtest)
print(np.round(predictx))
```

```python
#calculating mean_squared_error-it find the squared difference between actual vale and predicted value 
#r2_score measures how independent variable(hours) explain variability of dependent variable(score)
#r2=1:perfect prediction with no errors,r2=0:does not explain variability 

from sklearn.metrics import mean_squared_error,r2_score
y=mean_squared_error(ytest,predictx)
print('mean_squared_error = ',np.round(y))
z=r2_score(ytest,predictx)
print('r2_score = ',np.round(z))
```

```python
#plotting regression line 
#Each point on the line represents a predicted value of the dependent variable (score) 
#for a given value of the independent variable (hours)
plt.scatter(X,Y)
plt.plot(X,model.predict(X),color='red')
plt.title('Hours vs score (linear regression)')
plt.xlabel('Hours')
plt.ylabel('score')
```

```python
#comparing actual and predicted values 
c=pd.DataFrame({'actual':ytest,'predicted':np.round(predictx)})
print(c)
```

```python
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=UserWarning)
h=1
t=model.predict([[h]])
print("No of Hours =",h)
print("Predicted Score = ",np.round(t[0]))
```

```python
#evaluate the performance 
from sklearn.metrics import mean_absolute_error
q=mean_absolute_error(ytest,predictx)
print('mean_absolute_error = ',q)
```
